```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
n = 100
theta = 5 #выбрали
beta = 0.95

np.random.seed(42) #для воспроизводимости
X = np.random.uniform(theta, 2*theta, n)
print(X)
```

```python
mx = np.mean(X)
xmax = np.max(X)
theta1=2/3*mx #ОММ
theta2=(n+1)*np.max(X)/(2*n+1) #ОМП исправленный
mu2 = np.sum([(x_i - mx)**2 for x_i in X]) / n 
```

```python
left_accur = xmax /(1 + ((1+beta)/2) ** (1/n))
right_accur = xmax /(1 + ((1-beta)/2) ** (1/n))
l1 = right_accur - left_accur
print(f"Точный доверительный интервал: [{left_accur}, {right_accur}]")
print(f"Длина доверительного интервала: {l1}")
```

```python
t1 =  st.norm.ppf((1 - beta) / 2)
t2 = st.norm.ppf((1 + beta) / 2)
asymptotic_left = theta1 - 2/3 * np.sqrt(mu2/n) * t2
asymptotic_right = theta1 - 2/3 * np.sqrt(mu2/n) * t1
l2 = asymptotic_right - asymptotic_left
print(f"Асимптотический доверительный интервал: [{asymptotic_left}, {asymptotic_right}]")
print(f"Длина асимптотического доверительного интервала: {l2}")
```

Численно посторить бутстраповский доверительный интервал

```python
N = 1000
values = []
for _ in range(N):
    sample  = np.random.choice(X, size=n, replace=True)
    values.append(np.mean(sample) * 2/3 - theta1)

values = sorted(values)
t1 = int((1 - beta) / 2 * N-1) #индексы для квантилей
t2 = int((1 + beta) / 2 * N-1)
bootstrap_left = theta1 - values[t2]
bootstrap_right = theta1 - values[t1]
l3 = bootstrap_right - bootstrap_left
print("Для ОММ:")
print(f"Бутстраповский доверительный интервал: [{bootstrap_left}, {bootstrap_right}]")
print(f"Длина бутстраповского доверительного интервала: {l3}")
```

```python
N = 1000
values = []
for _ in range(N):
    sample  = np.random.choice(X, size=n, replace=True)
    values.append(np.max(sample) * (n+1)/(2*n+1) - theta2)

values = sorted(values)
t1 = int((1 - beta) / 2 * N-1) #индексы для квантилей
t2 = int((1 + beta) / 2 * N-1)
bootstrap_left = theta2 - values[t2]
bootstrap_right = theta2 - values[t1]
l4 = bootstrap_right - bootstrap_left
print("Для ОМП:")
print(f"Бутстраповский доверительный интервал: [{bootstrap_left}, {bootstrap_right}]")
print(f"Длина бутстраповского доверительного интервала: {l4}")
```

```python
print("Сравнение длин интервалов: ")
print(f"Точный: {l1}")
print(f"Асимптотический: {l2}")
print(f"Бутстраповский (ОММ): {l3}")
print(f"Бутстраповский (ОМП): {l4}")
```
