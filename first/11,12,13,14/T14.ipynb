```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import scipy.optimize as opt
import seaborn as sns
import math as mt
from skimage.io import imread, imsave

np.random.seed(42)
plt.style.use('ggplot')
```

```python
x = np.array([-1.11, -6.10, 2.42])
y = np.array([-2.29, -2.91])
sgm_x_sq = 2
sgm_y_sq = 1
alpha = 0.05
mx = np.mean(x)
my = np.mean(y)
n = len(x)
m = len(y)
delt = (mx - my) / np.sqrt((sgm_x_sq/n + sgm_y_sq/m))
crit = st.norm.ppf(1-alpha)
print("Статистика критерия:", delt, "Критическая область G: delt >", crit)
print("Нет веских оснований отвергнуть H0" if delt<crit else "Отвергаем H0")
```

```python
X = np.linspace(-1,5,1000)
quant = st.norm(0, np.sqrt(sgm_x_sq/n + sgm_y_sq/m)).ppf(1-alpha)
Y = 1 - st.norm(X, np.sqrt(sgm_x_sq/n + sgm_y_sq/m)).cdf(quant)
_, ax = plt.subplots(1,1, figsize=(10,5))
ax.plot(X, Y, color='green')
ax.set_title("График Мощности")
ax.set_xlabel("theta")
ax.set_ylabel("W(theta)")
plt.show()
```
