```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
theta = 5
n = 100
beta = 0.95

np.random.seed(42)
X = st.pareto(b=theta-1, scale=1).rvs(size=n)
X
```

```python
theta_omp = 1 + 1/ np.mean(np.log(X))
t1 = st.norm.ppf((1 - beta) / 2)
t2 = st.norm.ppf((1 + beta) / 2)

med_omp = 2 ** (1/(theta_omp-1))
med_theor = 2 ** (1/theta -1)

asymp_omp_left = med_omp - (med_omp * np.log(2)) / (np.sqrt(n) * (theta_omp-1)) * t2
asymp_omp_right = med_omp - (med_omp * np.log(2)) / (np.sqrt(n) * (theta_omp-1)) * t1
l1 = asymp_omp_right - asymp_omp_left
print(f"Асимптотический доверительный интервал для медианы ОМП: [{asymp_omp_left}, {asymp_omp_right}]")
print(f"Длина асимптотического доверительного интервала для медианы ОМП: {l1}")
```

```python
asymp_left = theta_omp - t2/(np.sum(np.log(X)) / np.sqrt(n))
asymp_right = theta_omp - t1/(np.sum(np.log(X)) / np.sqrt(n))
l2 = asymp_right - asymp_left
print(f"Асимптотический доверительный интервал для theta ОМП: [{asymp_left}, {asymp_right}]")
print(f"Длина асимптотического доверительного интервала для theta ОМП: {l2}")

```

```python
N = 1000
values = []

for _ in range(N):
    sample = np.random.choice(X, size=n, replace=True)
    values.append(1 + 1/ np.mean(np.log(sample)))
    
values = sorted(values)
t1 = int((1 - beta)/2 * (N-1))
t2 = int((1 + beta)/2 * (N-1))

left_boot = values[t1]
right_boot = values[t2]
l3 = right_boot - left_boot
print(f"Непараметрический Бутстраповский доверительный интервал для theta ОМП: [{left_boot}, {right_boot}]")
print(f"Длина непараметрического бутстраповского доверительного интервала для theta ОМП: {l3}")


```
